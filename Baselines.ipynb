```python
rf = RandomForestRegressor(max_depth = 100, random_state = 7)
reg = linear_model.LinearRegression()
```

```python
##RQ1 Linear regression

count = 0
for user in lst_rq1:
    dataframe = pd.read_pickle('Users_prepped/All/PerUser/{}_final'.format(str(user)))
    dataframe = dataframe[list(dataframe.columns)[1:]]
    dataset = dataframe.values
    dataset = dataset.astype('float32')
    
    #Scale Features
    scaler = MinMaxScaler(feature_range=(0, 1))
    arrX = dataset[:, :-1]
    arrY = dataset[:, -1]
    arrY = arrY.reshape(arrY.shape[0], 1)
    arrX = scaler.fit_transform(arrX)
    dataset = np.concatenate((arrX, arrY), axis = 1)
    
    trainX, testX, trainY, testY = train_test_split(dataset[:, :-1], dataset[:, -1], test_size=0.1, random_state=7)
    reg.fit(trainX, trainY)
    predY = reg.predict(testX)
    predY_train = reg.predict(trainX)
    
    trainrmse = mean_squared_error(trainY, predY_train, squared = False)
    rmse = mean_squared_error(testY, predY, squared = False)
    mae = mean_absolute_error(testY, predY)
    
    with open('Baseline_docu/RQ1_LinReg_Docu.txt', 'a') as docu_file:
        docu_file.write('{},{},{},{}\n'.format(user, trainrmse, rmse, mae))
        
    if count == 0:
        df_out = pd.DataFrame([user,trainrmse, rmse, mae]).T
        df_out.set_axis(['User','Train_RMSE', 'RMSE', 'MAE'], axis = 1, inplace = True)
        count += 1
    else:
        df_out = pd.read_pickle('Baseline_docu/RQ1_LinReg_df')
        df_out.loc[len(df_out)] = [user, trainrmse, rmse, mae]
    df_out.to_pickle('Baseline_docu/RQ1_LinReg_df')

print('RQ 1 Linear Regression')
print('Train RMSE: {}'.format(df_out['Train_RMSE'].mean()))
print('RMSE: {}'.format(df_out['RMSE'].mean()))
print('MAE: {}'.format(df_out['MAE'].mean()))

```
